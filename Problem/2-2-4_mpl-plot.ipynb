# ライブラリの`import`を用いる
## matplotlibを使ってグラフをプロット

まずはデータを読み込み，それをX, Yのデータにして返す関数を作りましょう。
csv とはcomma separated valuesです。テキストで開いてみればわかります。エクセルでも開けます。

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name, 'r'))
    X = [2]
    Y = [3]
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X, Y  # データを返す
read_data(file_name)
```

matplotlibを用いてデータを散布図にプロットする関数を作成します。

```python
import matplotlib.pyplot as plt
% matplotlib inline
#  % matplotlib inlineはjupyter notebookでmatplotlibを行内に表示するためのおまじない
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np
x = np.arange(-3,3,0.1)
y = np.sin(x)
plt.title("title")
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.plot(x,y)
plt.show() 
```

では，データを読み込んで描画して見ましょう。

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name, 'r'))
    X = [2]
    Y = [3]
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X, Y  # データを返す

import matplotlib.pyplot as plt
% matplotlib inline

def draw_graph(x,y,title,xlabel,ylabel):
    plt.title("title")
    plt.xlabel("X-LABEL")
    plt.ylabel("Y-LABEL")
    plt.plot(x,y)
    plt.show() 

file_name = 'data3.csv'  
x,y = read_data(file_name)
title,xl,yl = 'LineGraph','X-Axis','Y-Axis'
draw_graph(x, y, title, xl, yl) # 関数draw_graphによりグラフ描画
```
